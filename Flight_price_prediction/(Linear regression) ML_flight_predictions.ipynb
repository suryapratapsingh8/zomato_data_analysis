```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel("Data_Train.xlsx")
train_df.head()
```

```python
test_df = pd.read_excel("Test_set.xlsx")
test_df
```

```python
train_df.isnull().sum()
```

```python
train_df.shape
```

```python
train_df.dropna(inplace=True)
train_df
```

```python
train_df.isnull().sum()
```

```python
train_price = train_df['Price']
```

```python
train_df = train_df.drop('Price', axis=1)
```

```python
train_df['is_train'] = 1
test_df['is_train'] = 0
```

```python
combined_df = pd.concat([train_df, test_df], ignore_index=True)
```

```python
# 1. Date_of_Journey
combined_df['Date_of_Journey'] = pd.to_datetime(combined_df['Date_of_Journey'], format='%d/%m/%Y')
combined_df['Journey_Day'] = combined_df['Date_of_Journey'].dt.day
combined_df['Journey_Month'] = combined_df['Date_of_Journey'].dt.month
combined_df = combined_df.drop('Date_of_Journey', axis=1)
```

```python
combined_df
```

```python
# 2. Dep_Time
combined_df['Dep_Time'] = pd.to_datetime(combined_df['Dep_Time'])
combined_df['Dep_Hour'] = combined_df['Dep_Time'].dt.hour
combined_df['Dep_Minute'] = combined_df['Dep_Time'].dt.minute
combined_df = combined_df.drop('Dep_Time', axis=1)
```

```python
combined_df
```

```python
# 3. Arrival_Time
combined_df['Arrival_Time'] = pd.to_datetime(combined_df['Arrival_Time'], errors='coerce')
# Some arrival times might have date info (e.g., '01:10 22 Mar'), pandas handles this.
# Let's re-extract hour and minute
combined_df['Arrival_Hour'] = combined_df['Arrival_Time'].dt.hour
combined_df['Arrival_Minute'] = combined_df['Arrival_Time'].dt.minute
combined_df = combined_df.drop('Arrival_Time', axis=1)
```

```python
# 4. Duration
def convert_duration_to_minutes(duration):
    hours = 0
    minutes = 0
    if 'h' in duration:
        hours = int(duration.split('h')[0])
        if 'm' in duration:
            minutes = int(duration.split('h')[1].strip().split('m')[0])
    elif 'm' in duration:
        minutes = int(duration.split('m')[0])

    return (hours * 60) + minutes

combined_df['Duration_Total_Minutes'] = combined_df['Duration'].apply(convert_duration_to_minutes)
combined_df = combined_df.drop('Duration', axis=1)
```

```python
print("\nUnique values in 'Total_Stops':")
print(combined_df['Total_Stops'].unique())

# Create a mapping
stop_mapping = {
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4
}

combined_df['Total_Stops'] = combined_df['Total_Stops'].map(stop_mapping)
```

```python
combined_df['Total_Stops'].isnull().sum()
```

```python
# 6. Analyze Categorical Features
print("\nUnique values in 'Airline':")
print(combined_df['Airline'].unique())
print(f"Count: {combined_df['Airline'].nunique()}")

print("\nUnique values in 'Source':")
print(combined_df['Source'].unique())
print(f"Count: {combined_df['Source'].nunique()}")

print("\nUnique values in 'Destination':")
print(combined_df['Destination'].unique())
print(f"Count: {combined_df['Destination'].nunique()}")

print("\nValue counts for 'Additional_Info' (Top 10):")
print(combined_df['Additional_Info'].value_counts().head(10))

# Based on 'Additional_Info', over 78% is 'No info'.
# We can either group the less frequent ones or just encode 'No info' vs 'Other info'.
# Let's drop 'Route' as it's redundant with 'Total_Stops', 'Source', and 'Destination'.
combined_df = combined_df.drop('Route', axis=1)

# For 'Additional_Info', let's replace 'No info' with a more neutral 'Other'
# and then group all other specific infos into a single category.
# Actually, 'No info' is the dominant class. Let's just encode the top few and group the rest.
# Given that 'No info' is ~80%, this column is unlikely to be very predictive.
# Let's simplify and drop it for this model.
combined_df = combined_df.drop('Additional_Info', axis=1)

# Show the dataframe info after preprocessing
print("\nCombined Data Info after Preprocessing:")
combined_df.info()

# Show the head
print("\nCombined Data Head after Preprocessing:")
print(combined_df.head())
```

```python
combined_df.head()
```

```python
categorical_features = ['Airline', 'Source', 'Destination']
combined_df_encoded = pd.get_dummies(combined_df, columns=categorical_features, drop_first=True)
```

```python
combined_df_encoded.head()
```

```python
# Split back into train and test sets using the 'is_train' flag
X_train_final = combined_df_encoded[combined_df_encoded['is_train'] == 1].drop('is_train', axis=1)
X_test_final = combined_df_encoded[combined_df_encoded['is_train'] == 0].drop('is_train', axis=1)
y_train_final = train_price
```

```python
# Align columns - just in case any category was present in one set but not the other
# (though combining them first should have prevented this)
X_train_cols = set(X_train_final.columns)
X_test_cols = set(X_test_final.columns)

missing_in_test = list(X_train_cols - X_test_cols)
for col in missing_in_test:
    if col != 'Price': # 'Price' was never in the combined_df
        X_test_final[col] = 0

missing_in_train = list(X_test_cols - X_train_cols)
for col in missing_in_train:
    X_train_final[col] = 0

# Ensure the column order is the same
X_test_final = X_test_final[X_train_final.columns]
```

```python
# --- 3. Exploratory Data Analysis (EDA) ---
# We'll perform EDA on the *training* data (before splitting for validation)
# Let's add the 'Price' back to the X_train_final for easier plotting
eda_df = X_train_final.copy()
eda_df['Price'] = y_train_final.values
```

```python
# a. Distribution of Price
plt.figure(figsize=(10, 6))
sns.histplot(eda_df['Price'], kde=True, bins=50)
plt.title('Distribution of Flight Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.savefig('price_distribution.png')
plt.close()

plt.figure(figsize=(10, 6))
sns.histplot(np.log1p(eda_df['Price']), kde=True, bins=50)
plt.title('Distribution of Log-Transformed Flight Price')
plt.xlabel('Log(Price)')
plt.ylabel('Frequency')
plt.savefig('price_log_distribution.png')
plt.close()
print("\nGenerated price distribution plots.")
```

**LINEAR REGRESSION MODEL**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import pickle

# --- 1. Split Training Data ---
# We assume X_train_final and y_train_final are in memory from the previous run
try:
    X_train, X_val, y_train, y_val = train_test_split(X_train_final, y_train_final, test_size=0.3, random_state=42)
except NameError:
    print("Error: 'X_train_final' or 'y_train_final' not found in memory.")
    print("Please run the previous data processing and feature engineering steps first.")
    # Stop execution if variables aren't present
    exit()

print(f"New training set shape: X={X_train.shape}, y={y_train.shape}")
print(f"Validation set shape: X={X_val.shape}, y={y_val.shape}")

# --- 2. Scale Features ---
# Linear Regression is sensitive to feature scales, unlike Random Forest.
print("\nScaling features using StandardScaler...")
scaler = StandardScaler()

# Fit on training data and transform it
X_train_scaled = scaler.fit_transform(X_train)

# Transform validation data using the same scaler
X_val_scaled = scaler.transform(X_val)
print("Scaling complete.")

# --- 3. Model Selection and Training ---
print("\nTraining LinearRegression model...")
lr_model = LinearRegression()

# Train the model on the (80%) SCALED training split
lr_model.fit(X_train_scaled, y_train)
print("Model training complete.")

# --- 4. Model Evaluation ---
print("\nEvaluating model on scaled validation set...")
# Make predictions on the scaled validation set
y_pred_val_lr = lr_model.predict(X_val_scaled)

# Calculate and print metrics
r2_score_lr = metrics.r2_score(y_val, y_pred_val_lr)
mae_lr = metrics.mean_absolute_error(y_val, y_pred_val_lr)
rmse_lr = np.sqrt(metrics.mean_squared_error(y_val, y_pred_val_lr))

print(f"Linear Regression Validation Metrics:")
print(f"  R^2 Score: {r2_score_lr:.4f}")
print(f"  Mean Absolute Error (MAE): {mae_lr:.2f}")
print(f"  Root Mean Squared Error (RMSE): {rmse_lr:.2f}")

# --- 5. Final Training & Prediction ---
# Now, retrain the model on the *entire* training dataset, but first scale it.
print("\nScaling the entire training dataset for final model...")
final_scaler = StandardScaler()
X_train_final_scaled = final_scaler.fit_transform(X_train_final)

print("Retraining LinearRegression model on the entire scaled training dataset...")
lr_model_final = LinearRegression()
lr_model_final.fit(X_train_final_scaled, y_train_final)
print("Final model training complete.")

# Scale the actual test set (X_test_final) using the final_scaler
print("Scaling the test set...")
# We assume X_test_final is in memory
X_test_final_scaled = final_scaler.transform(X_test_final)

# Make predictions on the scaled test set
print("Making predictions on the test set...")
test_predictions_lr = lr_model_final.predict(X_test_final_scaled)

# --- 6. Output ---
# Create a DataFrame for the predictions
predictions_lr_df = pd.DataFrame(test_predictions_lr, columns=['Predicted_Price'])

# Save the predictions to a CSV file
predictions_lr_df.to_csv("flight_price_predictions_linear_regression.csv", index=False)

print(f"\nSuccessfully saved {len(test_predictions_lr)} predictions to 'flight_price_predictions_linear_regression.csv'.")
```
