To get the information about all the columns's datatypes.

**Integer (int):** Whole numbers without a fractional component (e.g., 1, -5, 0).

 **Floating-Point (Float/Double):** Numbers that have a fractional or decimal component (e.g., 3.14, -0.5).

**Boolean (bool):** Logical values representing only two states (e.g., True, False).

**String (str/text):** A sequence of characters representing text, including letters, digits, and symbols (e.g., "Apple", "ID-42").

**Character (char):** A single alphabetic letter, digit, or symbol (e.g., 'X', '9').

**Date:** A specific calendar point defined by year, month, and day (e.g., 2025-11-10).

**Time:** A specific point in the day defined by hours, minutes, and seconds (e.g., 14:30:00).

**Date/Time (Timestamp):** A precise moment combining date and time information (e.g., 2025-11-10 14:30:00).

**Categorical/Enum:** Values restricted to a small, predefined set of unique labels or states (e.g., "High", "Medium", "Low").

**Binary Data:** Raw sequences of bytes, often used for images, audio, or compressed files (e.g., image file data).

To get the mathematical description of data i.e **count, mean, standard deviation, min value, max value, 25%, 50% and 75%.**

**25% **(First Quartile or Q1)
Significance: This value is the 25th percentile.

What it means: 25% of the data points in that column fall at or below this value. The remaining 75% of the data is above it.

**50%** (Second Quartile or Median/Q2)
Significance: This value is the 50th percentile and is also known as the median.

What it means: 50% of the data points fall at or below this value, and the other 50% fall above it. It is the exact middle value of the dataset.

**75% **(Third Quartile or Q3)
Significance: This value is the 75th percentile.

What it means: 75% of the data points fall at or below this value. The smallest 25% of the data is above it.

```python
country_df = pd.read_excel('Country-Code.xlsx')
country_df.head()
```

As we can see there is country code in zomato dataset and here in counry code dataset also so in this case we will combine these two dataset in one and wrt country code we will have country name in zomato dataset.

```python
final_df = pd.merge(df, country_df, on = 'Country Code', how = 'left')
final_df
```

```python
# Name of coumtry in list type
country_name = final_df.Country.value_counts().index
```

```python
# Saving count of values wrt above country list type in list.
c_values = final_df.Country.value_counts().values
```

Plotting the chart.

```python
plt.pie(c_values[:3], labels= country_name[:3], autopct='%1.2f%%')
```

Maximum transaction: India >> US > UK

```python
final_df.columns
```

```python
rating = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0: 'rating count'})
rating
```

Observation:
1. Most people have not rated the app (2148)
2. 1.8 - 2.4 -> POOR
3. 2.5 - 3.4 -> AVERAGE
4. 3.5 - 3.9 -> GOOD
5. 4.0 - 4.4 -> VERY GOOD
6. 4.5 - 4.9 -> Excellent


```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10,6)
sns.barplot(x="Aggregate rating", y = "rating count", data = rating)
```

```python
# Finding name of countries that has given 0 rating
final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

```python
# Finding out which currency is used by which country
final_df.groupby(['Country', 'Currency']).size().reset_index()
```

```python
# Which country do have online deliveries
final_df[['Has Online delivery', 'Country']].groupby(['Has Online delivery', 'Country']).size().reset_index()
```

```python
label = final_df.City.value_counts().index
city = final_df.City.value_counts().values
```

```python
plt.pie(city[:5], labels= label[:5], autopct='%1.2f%%')
```

```python
# Finding top 10 cuisines
final_df[['Cuisines']].groupby('Cuisines').size().reset_index(name='Count').sort_values(by='Count', ascending=False).head(10)
```
